# Gradient Decent Method

#### Objective function

$ f(x) = 6x^2 + 9x + 3 $

```python
import numpy as np
import matplotlib.pyplot as plt

# IMPLEMENT HERE objective_function(x):
def objective_function(x):
    return 6 * x**2 + 9 * x + 3

x = np.arange(-5, 5, 0.1)
y = objective_function(x)


plt.plot(x, y)
plt.show()
```

#### Find minimum x using scikit_learn

```python
from scipy import optimize

# IMPLEMENT HERE
result = optimize.minimize(objective_function, x0=0)

plt.plot(x, y)
plt.plot(result.x, objective_function(result.x), 'ro')
plt.show()
```

#### Find minimum x using graident descent method

```python
from sre_constants import OPCODES
import numpy as np

def minimize(func, x0, tol=1e-16, max_iter=10000, alpha= 0.01) :

    eps = 1e-8
    x=x0
    for i in range(max_iter) :
      gradient = (func(x+eps) - func(x)) / eps
      x = x - alpha * gradient
      if abs(gradient) < tol : # Check the absolute value of the gradient
        break

    return x

result = minimize(objective_function, x0 = 5, alpha = 0.1)

plt.plot(x, y)
plt.plot(result, objective_function(result), 'ro')
plt.show()
```

#### Local minima Problem




```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function_2(x):
    return x**6 - 5*x**4 + 7*x**2 - 3

x = np.arange(-2, 2, 0.1)
y = objective_function_2(x)


plt.plot(x, y)
plt.show()
```
